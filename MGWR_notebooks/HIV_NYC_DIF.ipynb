```python
import pandas as pd

df=pd.read_csv(r'HIV_NYC_GWR.csv')
df = df.drop(columns = ['HIV_Case_log'])
df.head()
```

```python
df3 = df.drop(columns=['geoid', 'HIV_Case', 'HIV_Rate', 'X', 'Y'])
df3
```

```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
X = df3.iloc[:,:]
calc_vif(X)
```

```python
# remove'% health insurance'
df2 = df.drop(columns=['geoid', 'X', 'Y', 'HIV_Case', 'HIV_Rate', '%_health_insurance'])
df2
```

```python
X = df2.iloc[:,:]
calc_vif(X)
```

```python
import numpy as np

# %without health insurance
# %vacant housing
# %below poverty
# %unemployed
df_independent = df[['%_no_health_insurance', '%_below_poverty', '%_vacant_housing', '%_white']]
print(calc_vif(df_independent))

# correlation coefficient matrix
cc_matrix = pd.DataFrame(np.corrcoef(df_independent.transpose()))
cc_matrix.columns = df_independent.columns
cc_matrix.index = df_independent.columns
cc_matrix
```
